### 1. Problem

1) Using decision trees to classify the Iris dataset.

2) Calculate the accuracy and F1 score to evaluate the performance of the model.

3) Draw the decision tree generated by the model. Draw the confusion matrix to visualize the classification results on the test set.

### 2. Data processing

1) Load the Iris dataset.

2) Print feature names, and class names of the target variable.

3) Split the training set and the test set.


```python
# load packages
import numpy as np
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
X = iris.data
y = iris.target

# Display feature names
print("features::", iris.feature_names)
 
# Display labels (The class of the target variable)
print("labels:", iris.target_names)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.4)
```

### 3. Decision Tree classifier

```python
# Load Python packages for the decision tree
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
# Train the Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42) 
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

### 4. Permance metrics

```python
# Load Python packages to evaluate the performance
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

# 计算准确率、召回率、精确率和F1分数
accuracy = accuracy_score(y_test, y_pred) 
recall = recall_score(y_test, y_pred, average='micro') 
precision = precision_score(y_test, y_pred, average='micro') 
f1 = f1_score(y_test, y_pred, average='micro') 

print("Accuracy:", accuracy) 
print("Recall:", recall) 
print("Precision:", precision) 
print("F1 Score:", f1)
```

### 4. Visualization
1) Draw the decision tree generated by the model.

2) Draw the confusion matrix to visualize the classification results on the test set.

```python
# Load Python packages for visualization
import matplotlib.pyplot as plt  
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib as mpl
from sklearn import tree

mpl.rcParams["font.size"] = 24
```

```python
fig, ax = plt.subplots(figsize=(9, 9))
_ = ConfusionMatrixDisplay.from_estimator(
    clf, X_test, y_test, 
    display_labels=iris.target_names,  # 使用多分类的标签
    ax=ax, 
    cmap='Blues'
)
```

```python
# Plot the decision tree
fig, ax = plt.subplots(figsize=(12, 12))
_ = tree.plot_tree(
    clf,
    feature_names=iris.feature_names,  # 使用正确的特征名称
    class_names=iris.target_names,     # 使用正确的分类名称
    filled=True,
    ax=ax,
    proportion=True,
    node_ids=True,
    rounded=True
)

plt.show()
```
