```python
def iter(z,c,a,m):   return (a*z+c) % m
def run(n,z0=27,c=47,a=8,m=100):
    seq_list = []
    num_list = []
    z = z0
    for _ in range(int(n)):
        seq = [z]
        num = z*10000

        z = iter(z,c,a,m)
        seq.append(z)
        num = num +z*100

        z = iter(z,c,a,m)
        seq.append(z)
        num = num + z

        z = iter(z,c,a,m)

        if num in num_list:
            break
        seq_list.append(seq)
        num_list.append(num)
    return seq_list,num_list

seq,num = run(int(1e5))
print(len(num))
print(num)

seq,num = run(int(1e4),z0=0)
print(len(num))
print(num)

seq,num = run(int(1e5),z0=0,c=0)
print(len(num))
print(num)
```

```python
def run(n,z0=1,c=13,a=13,m=16):
    z_list = []
    z = z0
    for _ in range(int(n)):
        if z in z_list:
            break
        z_list.append(z)
        z = (a*z+c) % m

    return z_list
```

```python
zl1 = run(int(1e5),z0=1)
print(len(zl1))
print(zl1)

zl2 = run(int(1e5),z0=1,a=12)
print(len(zl2))
print(zl2)


zl3 = run(int(1e5),z0=1,c=12)
print(len(zl3))
print(zl3)
```
